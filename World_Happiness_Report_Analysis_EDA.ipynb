Finding Outliers by IQR

```python
sns.boxplot(x="Happiness_score",data=df1)
```

```python
Q1 = df1['Happiness_score'].quantile(0.25)
print(Q1)
```

```python
Q3 = df1['Happiness_score'].quantile(0.75)
print(Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
print(lower_bound)

upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
```

```python
# Finding the Outliers (Masking)
outliers = df1[(df1['Happiness_score'] < lower_bound) | (df1['Happiness_score'] > upper_bound)]
outliers
```

```python
from IPython.display import display
display(outliers)
```

```python
df2=df.copy()
```

```python
df2 = df1[(df1['Happiness_score'] >= lower_bound) & (df1['Happiness_score'] <= upper_bound)]
```

A horizontal bar chart representing the Top 10 Happiest Countries based on their Happiness Scores.

```python
top_10_countries = df2.head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='Happiness_score', y='Country name', data=top_10_countries, hue='Country name', palette='Blues_r', legend=False)

# Customize plot
plt.title("Top 10 Happiest Countries", fontsize=14)
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.xlim(0, top_10_countries['Happiness_score'].max() + 0.5)

plt.show()
```

A horizontal bar chart representing the Happiness Scores of the bottom 10 countries

```python
last_countries_10 = df2.tail(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='Happiness_score', y='Country name', data=last_countries_10, hue='Country name', palette='Reds', legend=False)

plt.title("Top 10 Happiest Countries", fontsize=14)
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.xlim(0, top_10_countries['Happiness_score'].max() + 0.5)

plt.show()
```

This scatter plot helps us understand the relationship between a country's economic wealth (Log GDP per capita) and happiness levels (Happiness Score).

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df2, x='Log GDP per capita', y='Happiness_score', hue='Country name', alpha=0.7, palette="viridis")

plt.title("Log GDP per Capita vs. Happiness Score", fontsize=14)
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend([],[], frameon=False)

plt.show()
```

This regression plot helps us analyze the relationship between Social Support and Happiness Score for different countries.

```python
plt.figure(figsize=(10, 6))

sns.regplot(data=df2, x='Social support', y='Happiness_score', scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})

plt.title("Social Support vs. Happiness Score", fontsize=14)
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")

plt.show()
```

This is a bar plot comparing the Social Support and Happiness Score for the top 10 countries

```python
top_10_countries = df2.nlargest(10, 'Social support')

plt.figure(figsize=(12, 6))


bar_width = 0.4


positions = range(len(top_10_countries))


plt.bar(positions, top_10_countries['Social support'], width=bar_width, label="Social Support", color='royalblue')

plt.bar([p + bar_width for p in positions], top_10_countries['Happiness_score'], width=bar_width, label="Happiness Score", color='lightcoral')

plt.xticks([p + bar_width/2 for p in positions], top_10_countries['Country name'], rotation=45, ha='right')
plt.title("Top 10 Countries: Social Support vs Happiness Score", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Score")
plt.legend()


plt.show()
```

This bar plot compares Log GDP per Capita and Happiness Score for the top 10 countries

```python
top_10_countries = df2.nlargest(10, 'Log GDP per capita')

plt.figure(figsize=(12, 6))

bar_width = 0.4

positions = range(len(top_10_countries))

plt.bar(positions, top_10_countries['Log GDP per capita'], width=bar_width, label="Log GDP per Capita", color='seagreen')

plt.bar([p + bar_width for p in positions], top_10_countries['Happiness_score'], width=bar_width, label="Happiness Score", color='goldenrod')

plt.xticks([p + bar_width/2 for p in positions], top_10_countries['Country name'], rotation=45, ha='right')
plt.title("Top 10 Countries: GDP per Capita vs Happiness Score", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Score")
plt.legend()

plt.show()
```
